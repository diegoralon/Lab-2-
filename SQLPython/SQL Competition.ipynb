```python
import numpy as np
import pandas as pd
heroes = pd.read_csv("heroes_information.csv")
```

```python
###instalar desde anaconda navigator######
from pandasql import *
```

```python
pysqldf = lambda q: sqldf(q,globals())
```

```python
query2 = " SELECT * FROM heroes limit 10"
```

```python
pysqldf(query2)
```

```python
# Seleccionando Columnas Key word SELECT 
# SELECT field FROM table 
names = " SELECT name FROM heroes "
pysqldf(names)
#RESPUESTA A-BOMB
```

```python
# SELECT DISTINCT , SELECT COUNT()
publishers = " SELECT DISTINCT Publisher FROM heroes "
pysqldf(publishers)
# Repuesta 24
```

```python
# Filtrar los resultados de mis caracteristicas con WHERE nos 
# deinfe un campo 
altura = " SELECT Height FROM heroes WHERE Height > '200' "
pysqldf(altura)
# RESPUESTA 59
```

```python
# WHERE AND cuando los dos elementos son verdaderos 
# aqui donde el campo 1 sea igual a 2 y el otro igual a
# otra cosa la 4 y 5 

hm = " SELECT Race FROM heroes WHERE Race = 'Human' AND Height > '200' " 
pysqldf(hm)
# RESPUESTA 12
```

```python
# BETWEEN 
lb = " SELECT Weight FROM heroes WHERE Weight BETWEEN 100 AND 200 "
pysqldf(lb)
# RESPUESTA 98
```

```python
# OR selecciona filas por condiciones pero no es neces
#ario que todas se cumplan vamos a tener un resultadon 
# SE METE EN el de arriba pero se agrega or para busquedas mas locas 
oj = " SELECT COUNT(*) FROM heroes WHERE `Eye color` IN ('blue', 'red') "
pysqldf(oj)
# RESPUESTA 271
```

```python
# BETWEEN / IN Funcionan igual donde el campo cumpla algunos de los elementos
hmm = " SELECT COUNT(Race) FROM heroes WHERE (Race IN ( 'Human', 'Mutant') AND `Hair color` = 'Green') OR (Race = 'Vampire' AND `Hair color` = 'Black') "
pysqldf(hmm)

# RESPUESTA 5             
```

```python
# Order By ordena en datos en especifico 
byu = " SELECT * FROM heroes ORDER BY Race DESC"
pysqldf(byu)
# RESPUESTA SOLOMON GRUNDY 
```

```python
# GROUP BY 
gb = " SELECT Gender, COUNT(Gender) FROM heroes GROUP BY (Gender) "
pysqldf(gb)
# RESPUESTA FEMALE 200 Y MALE 505 
```

```python
# HAVING Porque no ahce una agregación 
phs = " SELECT Publisher FROM heroes GROUP BY (Publisher) HAVING COUNT(Publisher) >15 "
pysqldf(phs)
# RESPUESTA 4 
```
